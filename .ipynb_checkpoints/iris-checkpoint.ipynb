Question: Is this a balanced dataset vs imbalanced dataset ?



```python
# I get the number of observations for each species
repartition = iris["species"].value_counts().tolist()

# I want to put it in a Pandas Dataframe to display a pie chart
# I put in my a dictionary
dict_repart = {'nb_obs' : repartition}
# We put it in a form of Pandas Dataframe
df_class = pd.DataFrame(dict_repart, columns=['nb_obs'], index = ['Setosa', 'Versicolor', 'Viriginica'])


from pandas.plotting import table

plt.figure(figsize=(16,8))

# On the left, we plot a table telling us how many observation per species
ax1 = plt.subplot(121)
plt.axis('off')
tbl = table(ax1, df_class, loc='center')
tbl.auto_set_font_size(True)

# On the right, we plot a pie chart telling us the distribution of observation per species in percentage
ax2 = plt.subplot(122, aspect='equal')
plt.axis('off')
df_class.plot.pie(y='nb_obs',figsize=(15, 7),autopct='%1.1f%%', startangle=90, ax=ax2)
# tbl = table(ax1, df_class, loc='center')

plt.show()
```

We can see here that we have a <span style="color:green">balanced dataset</span>, since the number of data points for every class is 50.

## 3. Exploratory analysis

Here, we will be looking to answer questions such as:

* Are there any correlations in my data?

* Are there any confounding factors that explain these correlations?


**Rappel de la question de départ**: which of my 4 variable is (or are) more useful than other variables to distinguish setosa, verginica or versicolor?

**Our strategy**:


In this dataset, we have 3 features. So, it can be represented in 3 dimension.
Let's try to see if this view let us know if one of the features makes it possible to distinguish the classes.
So let's do a 3D plot to a get a general view and get the first conclusions.

If we have to choose 2 features, we are going to do bivariate analysis. So, a pair plot will give us a better representation and can give us some insights.

However, if we have to choose only one features, we will make univariate analysis using PDF and density plot to make a ranking of all our features from the most to least differentiating.

## <span style="color:blue;">3. 3D plot<span><a name="introduction"></a>

```python
import plotly.express as px
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
fig = px.scatter_3d(iris, x='sepal_length', y='sepal_width', z='petal_width',color='species')
fig.show()
```

After some mouse manipulations of this 3D graph, it seems that the Setosa species is linearly separable from the 2 species Versicolor and Virginica.

So that means they are well separated from Versicolor and Virginica.

On the other hand, the latter two species are overlapping. But it is hard to tell which of the features allows to better separate the 3 species.  

Let's check that with some pair plots. We will probably get a better view. Here we have only 4 features, so we will get 6 unique plots since 4C2 = 6

## <span style="color:blue;">4. Pair plots<span><a name="introduction"></a>

A pairs plot allows us to see both distribution of single variables and relationships between two variables.

Let's display the scatterplot matrix that we used earlier.

```python
sns.pairplot(iris)
;
```

Our data is normally distributed for the most part, which is great news if we plan on using any modeling methods that assume the data is normally distributed.

There's something strange going on with the petal measurements. Maybe it's something to do with the different Iris types. Let's color code the data by the class again to see if that clears things up.


```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(iris, hue="species", height=3, palette="husl", corner=True, markers=["o", "s", "D"]);
plt.show()
```

Sure enough, the strange distribution of the petal measurements exist because of the different species. This is actually great news for our classification task since it means that the petal measurements will make it easy to distinguish between Iris-setosa and the other Iris types.

Distinguishing Iris-versicolor and Iris-virginica will prove more difficult given how much their measurements overlap.

There are also correlations between petal length and petal width, as well as sepal length and sepal width. The field biologists assure us that this is to be expected: Longer flower petals also tend to be wider, and the same applies for sepals.


Observation :

* What we suspected with the 3d graph is true, namely that Setosa can be easily identified (linearly separable). And that Versicolor and Virginica have some overlap (almost linearly separable).

    if we have to rank the 4 features from most to least differentiating, we would go with:  PL >> PW > SL >> SW

* What we see here with pair plot is that <span style="color:green">petal_length and petal_width are the most useful features to identify various flower types</span>.

* Actually, we can find “lines” and “if-else” conditions to build a simple model to classify the flower types.

Now, that we have identified 2 features that can identify various flower type, it would be interesting to confirm or infirm that. We are going to make a unviariate analysis of all our features by plotting densities 2D plot. So that we can compare them and get a better insight.

## <span style="color:blue;">5. Univariate analysis : 2D density plots and PDF<span><a name="introduction"></a>

```python
# First, we put all flowers belonging to a species into a new Dataframe
iris_setosa = iris.loc[iris["species"] == "setosa"];
iris_virginica = iris.loc[iris["species"] == "virginica"];
iris_versicolor = iris.loc[iris["species"] == "versicolor"];

# we saw that petal_length and petal_width are particularly interesting for us, so let's write a distribution of PL and PW
sns.FacetGrid(iris, hue="species", height=5) \
   .map(sns.distplot, "petal_length") \
   .add_legend();

sns.FacetGrid(iris, hue="species", height=5) \
   .map(sns.distplot, "petal_width") \
   .add_legend();

# we will also plot histogram and PDF of sepal_width and sepal_width
sns.FacetGrid(iris,hue="species",height=5) \
    .map(sns.distplot,"sepal_width") \
    .add_legend();

sns.FacetGrid(iris,hue="species",height=5) \
    .map(sns.distplot,"sepal_length") \
    .add_legend();

plt.show();
```

**What we see above:**

2d histogram and PDF above allow us to determine the number of Setosa flowers that can be found with a certain petal length.


The line that we see above, are the Probability Density Function (PDF). This line smoothes the histogram / density plot(using a KDE).

**Observation:**

On the first and second plot (respectively, PL and PW): the setosa are better separable than using petal width. There is overlap between vericolor and virginca

On the third plot, we can see virginica and versicolor are fully overlapped.

On the last plot, we can’t separate any class because all of them are overlapped.

**Conclusion:**

We can detemrine a ranking of the better features to distinguish between flower species: 
 PL >> PW > SL >> SW
